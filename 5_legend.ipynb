```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 100, 30)
x
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))
plt.legend(['sin', 'cos', 'tan'])
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))
plt.legend(['sin', 'cos', 'tan'], loc = 'center') # default loc = 'best'
plt.show()
```
