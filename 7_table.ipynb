```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
rows = ['2011', '2012', '2013', '2014', '2015']
columns = ['7Ah', '35Ah', '40Ah', '135Ah', '150Ah']
data = [
    [75, 144, 114, 102, 108],
    [90, 126, 102 , 84, 126],
    [96, 104, 75, 105, 135],
    [105, 90, 150 ,90, 75],
    [90, 75, 135, 75, 90]
]
```

```python
colors = plt.cm.summer(np.linspace(0, 0.5, len(rows)))
colors
```

```python
index = np.arange(len(columns))
index
```

```python
index = np.arange(len(columns)) + 0.3
index
```

```python
bar_width = 0.3
```

```python
y_offset = np.zeros(len(columns))
y_offset
```

```python
fig = plt.figure()
ax = plt.subplot()
```

```python
cell_text = []
```

```python
n_rows = len(data)
n_rows
```

```python
ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row])

plt.show()
```

```python
ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width)
    y_offset += data[row]
plt.show()
```

```python
y_offset = np.zeros(len(columns))

ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width, bottom=y_offset)
    y_offset += data[row]
plt.show()
```

```python
y_offset = np.zeros(len(columns))

ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width, bottom=y_offset, color = colors[row])
    y_offset += data[row]
plt.show()
```

```python
y_offset = np.zeros(len(columns))

ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width, bottom=y_offset, color = colors[row])
    y_offset += data[row]
    cell_text.append(['%1.1f' % x for x in y_offset])

plt.show()
```

```python
y_offset = np.zeros(len(columns))
cell_text = []

ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width, bottom=y_offset, color = colors[row])
    y_offset += data[row]
    cell_text.append(['%1.1f' % x for x in y_offset])

the_table = ax.table(cellText = cell_text, rowLabels = rows, colLabels = columns, rowColours = colors, loc='top')

plt.show()
```

```python
y_offset = np.zeros(len(columns))
cell_text = []
bar_width = 0.5

ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width, bottom=y_offset, color = colors[row])
    y_offset += data[row]
    cell_text.append(['%1.1f' % x for x in y_offset])

the_table = ax.table(cellText = cell_text, rowLabels = rows, colLabels = columns, rowColours = colors, loc='bottom')

plt.xticks([])
plt.ylabel('Units sold')
plt.title('Number of batteries')
plt.show()
```

```python
y_offset = np.zeros(len(columns))
cell_text = []
bar_width = 0.5

ax = plt.subplot()

for row in range(n_rows):
    plot = ax.bar(index, data[row], bar_width, bottom=y_offset, color = colors[row])
    y_offset += data[row]
    cell_text.append(['%1.1f' % x for x in y_offset])

    i = 0
    for rect in plot:
        ax.text(rect.get_x() + rect.get_width() / 2, y_offset[i], '%d' % int(y_offset[i]), ha = 'center', va = 'bottom')
        i +=1
the_table = ax.table(cellText = cell_text, rowLabels = rows, colLabels = columns, rowColours = colors, loc='bottom')

plt.xticks([])
plt.ylabel('Units sold')
plt.title('Number of batteries')
plt.show()
```
