```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')

plt.show()
```

```python
a = np.arange(-5, 5, 0.2)
b = np.arange(-5, 5, 0.2)
```

```python
x, y = np.meshgrid(a, b) # make each array to 2 dimensional array with meshing
```

```python
z = np.cos(x) * np.sin(y)
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')

surf = ax.plot_surface(x, y ,z)

plt.show()
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')

surf = ax.plot_surface(x, y , z, cmap = plt.cm.summer)

plt.show()
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')

surf = ax.plot_surface(x, y , z, cmap = plt.cm.summer)

plt.colorbar(surf)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')

surf = ax.plot_surface(x, y , z, cmap = plt.cm.summer)

plt.colorbar(surf, shrink=0.5)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')

z = np.tan(x) * np.sin(y)

surf = ax.plot_surface(x, y , z, cmap = plt.cm.summer)

plt.colorbar(surf, shrink=0.5)
plt.show()
```
