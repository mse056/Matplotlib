```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
depts = ['CDGS', 'IT', 'Payroll', 'R&D', 'Sales & Marketing']
rp = [30, 15, 25, 10, 20, 30] # last number repeated of first number because of ploting
ra = [32, 20, 23, 11, 14, 32] # last number repeated of first number because of ploting
```

```python
plt.figure(figsize=(6, 10))
plt.subplot(projection='polar')
plt.show()
```

```python
theta = np.linspace(0, 2*np.pi, len(rp))
theta
```

```python
plt.figure(figsize=(6, 10))
plt.subplot(projection='polar')

lines, labels = plt.thetagrids(range(0, 360, int(360/len(depts))), depts)

plt.show()
```

```python
plt.figure(figsize=(6, 10))
plt.subplot(projection='polar')

lines, labels = plt.thetagrids(range(0, 360, int(360/len(depts))), depts)

plt.plot(theta, rp)
plt.plot(theta, ra)

plt.show()
```

```python
plt.figure(figsize=(6, 10))
plt.subplot(projection='polar')

lines, labels = plt.thetagrids(range(0, 360, int(360/len(depts))), depts)

plt.plot(theta, rp, label = 'Plan')
plt.fill(theta, rp, 'b', alpha=0.1) # alpha gives color transparency value
plt.plot(theta, ra, label = 'Actual')

plt.legend()
plt.title('Plan vs Actual spend any Department')
plt.show()
```
