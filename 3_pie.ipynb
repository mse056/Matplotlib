```python
from matplotlib import pyplot as plt
```

```python
market_share = [43, 32 , 21, 19]
companies = ['apple', 'samsung', 'sony', 'other']
```

```python
plt.pie(market_share, labels=companies)
plt.show()
```

```python
plt.pie(market_share, labels=companies, explode=[0, 0.1, 0, 0])
plt.show()
```

```python
plt.pie(market_share, labels=companies, explode=[0, 0.1, 0, 0], colors=['b', 'k', 'g', 'y'])
plt.show()
```

```python
plt.pie(market_share, labels=companies, radius=0.7) # default radius is 1
plt.show()
```

```python
plt.pie(market_share, labels=companies, autopct='%1f%%')
plt.show()
```

```python
plt.pie(market_share, labels=companies, autopct='%1.0f%%')
plt.show()
```

```python
plt.pie(market_share, labels=companies, labeldistance=0.5) # default labeldistance is 1.1
plt.show()
```

```python
plt.pie(market_share, labels=companies, startangle=90) 
plt.show()
```

```python
plt.pie(market_share, labels=companies, startangle=90, counterclock=False) 
plt.show()
```

```python
plt.pie(market_share)
plt.legend(companies)
plt.show()
```
