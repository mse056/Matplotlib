```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.arange(0, 100, 30)
data
```

```python
fig = plt.figure()

ax1 = fig.add_subplot()
ax1.plot()

plt.show()
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot()
ax1.plot()

plt.show()
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot()
ax1.plot(data, np.sin(data))

plt.show()
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot()
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot()
ax2.plot(data, np.cos(data))

plt.show() # ax1 is under the ax2
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(data, np.cos(data))

plt.show()
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(data, np.cos(data))

plt.show()
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot(3, 1, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(3, 1, 2)
ax2.plot(data, np.cos(data))

ax3 = fig.add_subplot(3, 1, 3)
ax3.plot(data, np.tan(data))

plt.show()
```

```python
fig = plt.figure(facecolor='y')

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(data, np.cos(data))

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(data, np.tan(data))

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8), facecolor='y') # default figsize = (6.4, 4.8)

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(data, np.cos(data))

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(data, np.tan(data))

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8), facecolor='y') # default figsize = (6.4, 4.8)

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(data, np.cos(data))

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(data, np.tan(data))

fig.subplots_adjust(wspace=1)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8), facecolor='y') # default figsize = (6.4, 4.8)

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(data, np.sin(data))

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(data, np.cos(data))

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(data, np.tan(data))

fig.subplots_adjust(wspace=1, hspace=1)

plt.show()
```
