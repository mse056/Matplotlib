```python
fig = plt.figure()
ax = fig.add_subplot()

verts = [(-1.5, 0), (0, 1), (1.5, 0), (0, -1), (-1.5, 0)]
codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
pat = Path(verts, codes) # draw the path

patch = patches.PathPatch(pat, lw=4) # patch the path

ax.add_patch(patch)

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

verts = [(-1.5, 0), (0, 1), (1.5, 0), (0, -1), (-1.5, 0)]
codes = [Path.MOVETO, Path.CURVE3, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
pat = Path(verts, codes) 

patch = patches.PathPatch(pat, lw=4) # lw means line width

ax.add_patch(patch)

ax.set_xlim(-2,2)
ax.set_ylim(-2,2)

plt.show()
```
