```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
salaries = [63000, 65000, 68000, 70000, 69000, 71000, 73000, 76000, 80000, 84000]
prices = [40000, 45000, 43000, 52000, 55000, 53000, 60000, 64000 , 62000, 59000]
```

```python
plt.bar(years, salaries)
plt.show()
```

```python
plt.bar(years, salaries)
plt.plot(years, prices, color = 'y')
plt.show()
```

```python
plt.bar(years, salaries)
plt.bar(years, prices)
plt.show()
```

```python
plt.bar(years, salaries)
plt.bar(years, prices)
plt.xticks(years, ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
plt.show()
```

```python
plt.bar(years, salaries, width = 0.3)
plt.bar(years, prices)
plt.xticks(years, ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
plt.show()
```

```python
x = np.arange(len(years))
w = 0.3

plt.bar(x, salaries, width = w)
plt.bar(x - w , prices, width = w)
plt.xticks(x, years)
plt.show()
```

```python
x = np.arange(len(years))
w = 0.3

plt.bar(x, salaries, width = w)
plt.bar(x - w , prices, width = w)
plt.bar(x + w , prices, width = w)
plt.xticks(x, years)
plt.show()
```

```python
plt.barh(years, salaries, color = 'y')
plt.show()
```

```python
plt.barh(years, salaries, color = 'y', edgecolor = 'k')
plt.show()
```
