```python
from matplotlib import pyplot as plt
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
salaries = [63000, 65000, 68000, 70000, 69000, 71000, 73000, 76000, 80000, 84000]
prices = [40000, 45000, 43000, 52000, 55000, 53000, 60000, 64000 , 62000, 59000]
```

```python
plt.plot(years)
plt.show()
```

```python
plt.plot(years, salaries)
plt.show()
```

```python
plt.plot(years, salaries)
plt.plot(years, prices)
plt.show()
```

```python
plt.plot(years, salaries)
plt.plot(years, prices)
plt.title('salaries vs prices')
plt.show()
```

```python
plt.plot(years, salaries)
plt.plot(years, prices)
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.show()
```

```python
plt.plot(years, salaries)
plt.plot(years, prices)
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend(['salareis', 'prices'])
plt.show()
```

```python
plt.plot(years, salaries, label = 'salaries')
plt.plot(years, prices, label = 'prices')
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend()
plt.show()
```

```python
plt.plot(years, salaries, label = 'salaries')
plt.plot(years, prices, label = 'prices')
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(years, salaries, 'o--y', label = 'salaries')
plt.plot(years, prices, label = 'prices')
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(years, salaries, 'o--y', label = 'salaries')
plt.plot(years, prices, label = 'prices', color = 'g', linestyle = '-.', marker = 'D')
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(years, salaries, 'o--y', label = 'salaries', linewidth = 3)
plt.plot(years, prices, label = 'prices', color = 'g', linestyle = '-.', marker = 'D')
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(years, salaries, 'o--y', label = 'salaries')
plt.plot(years, prices, label = 'prices', color = 'g', linestyle = '-.', marker = 'D')
plt.title('salaries vs prices')
plt.xlabel('years')
plt.ylabel('USD')
plt.legend()
plt.grid()
plt.savefig('one.png')
plt.show()
```
