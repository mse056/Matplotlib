```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 20 , 1000)
x
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1))

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1),xytext=(10,4))

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1),xytext=(10,4), arrowprops={'facecolor':'y'})

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1),xytext=(10,4), arrowprops={'facecolor':'y', 'shrink':0.05})

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1), xytext=(10,4), arrowprops={'facecolor':'y', 'shrink':0.05})
ax.annotate('minimum', xy=(5 * np.pi, -1), xytext=(2,-6), arrowprops={'arrowstyle':'->'})

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1), xytext=(10,4), arrowprops={'facecolor':'y', 'shrink':0.05})
ax.annotate('minimum', xy=(5 * np.pi, -1), xytext=(2,-6), arrowprops={'arrowstyle':'->', 'connectionstyle':'angle'})

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1), xytext=(10,4), arrowprops={'facecolor':'y', 'shrink':0.05})
ax.annotate('minimum', xy=(5 * np.pi, -1), xytext=(2,-6), arrowprops={'arrowstyle':'->', 'connectionstyle':'angle, rad=90'})

plt.axis('equal')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, np.cos(x))

ax.annotate('maximum', xy=(6.28, 1), xytext=(10,4), arrowprops={'facecolor':'y', 'shrink':0.05})
ax.annotate('minimum', xy=(5 * np.pi, -1), xytext=(2,-6), arrowprops={'arrowstyle':'->', 'connecationstyle':'angle, angleA=0, angleB=90'})

plt.axis('equal')
plt.show()
```
