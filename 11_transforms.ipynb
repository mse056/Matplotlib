```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
d = np.arange(0, 100, 10)
d
```

```python
fig = plt.figure(facecolor='y')
ax = fig.add_subplot()

ax.plot(d)

plt.grid()
plt.show()
```

```python
fig = plt.figure(facecolor='y')
ax = fig.add_subplot()

ax.plot(d)

ax.text(2, 50, 'middle of graph')

plt.grid()
plt.show()
```

```python
d = np.arange(0, 400, 10)

fig = plt.figure(facecolor='y')
ax = fig.add_subplot()

ax.plot(d)

ax.text(2, 50, 'middle of graph')

plt.grid()
plt.show()
```

```python
d = np.arange(0, 400, 10)

fig = plt.figure(facecolor='y')
ax = fig.add_subplot()

ax.plot(d)

ax.text(0.5, 0.5, 'middle of graph', transform = ax.transAxes)

plt.grid()
plt.show()
```
