```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
year = [1950,1960,1970,1980,1990,2000,2010,2018]
population_by_continent = {
    'africa':[228,284,365,477,631,814,1044,1275],
    'amercia':[340,425,519,619,727,840,943,1006],
    'asia':[1394,1686,2120,2625,3202,3714,4169,4560],
    'europe':[220,253,276,295,310,303,294,293],
    'oceania':[12,15,19,22,26,31,36,39]
}
```

```python
plt.stackplot(year, population_by_continent.values())
plt.show()
```

```python
plt.stackplot(year, population_by_continent.values(), labels=population_by_continent.keys())
plt.legend()
plt.show()
```

```python
x = np.random.randint(140, 200, 500)
plt.hist(x)
plt.show()
```

```python
plt.hist(x, bins=8)
plt.show()
```

```python
b = (140, 150, 160, 170, 180, 190, 200)
plt.hist(x, bins=b)
plt.show()
```

```python
b = (140, 150, 160, 170, 180, 190, 200)
plt.hist(x, bins=b, edgecolor = 'k')
plt.show()
```

```python
b = (140, 150, 160, 170, 180, 190, 200)
plt.hist(x, bins=b, edgecolor = 'k', orientation='horizontal')
plt.show()
```

```python
y = np.random.randint(140 , 200 , 300)

plt.hist([x, y], bins=b, edgecolor = 'k')
plt.show()
```

```python
plt.hist([x, y], bins=b, edgecolor = 'k', histtype='barstacked', color=['b','r'])
plt.show()
```

```python
countries = ['Germany', 'Turkey','vietnam','Egypt', 'philippines', 'Japan', 'Mexico', 'Russia', 'Bangladesh', 'Brazil']
population = [84278625, 84680273, 99460000, 104356852, 112992166, 124770000, 128665641, 146980061, 165158616, 215681000]
sizes = [842, 846,994,1043,1129,1247,1286,1469,1652,2156]
```

```python
plt.scatter(population, countries)
plt.show()
```

```python
plt.scatter(population, countries , s = 200)
plt.show()
```

```python
plt.scatter(population, countries , s = sizes)
plt.show()
```

```python
plt.scatter(population, countries , s = sizes, c = sizes)
plt.show()
```

```python
plt.scatter(population, countries , s = sizes, c = sizes, cmap='summer')
plt.show()
```

```python
plt.scatter(population, countries , s = sizes, c = sizes, cmap='summer')
plt.colorbar()
plt.show()
```

```python
plt.scatter(population, countries , s = sizes, c = sizes, cmap='summer', alpha=0.5) # default alpha = 1
plt.colorbar()
plt.show()
```

```python
plt.scatter(population, countries , s = sizes, c = sizes, cmap='summer', alpha=0.5, edgecolors='y')
plt.colorbar()
plt.show()
```

```python
plt.scatter(population, countries , s = sizes, c = sizes, cmap='summer', alpha=0.5, edgecolors='y', marker='p')
plt.colorbar()
plt.show()
```
