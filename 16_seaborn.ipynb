```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips') # loads tips from github repository which has files to test
tips
```

```python
sns.relplot(tips, x='total_bill', y='tip')

plt.show()
```

```python
sns.relplot(tips, x='total_bill', y='tip', col='time')

plt.show()
```

```python
sns.relplot(tips, x='total_bill', y='tip', col='time', hue='smoker')

plt.show()
```

```python
sns.relplot(tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker')

plt.show()
```

```python
sns.relplot(tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')

plt.show()
```

```python
sns.set_theme(context='poster')

sns.relplot(tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')

plt.show()
```

```python
sns.set_theme(context='poster', style='whitegrid') # style gives [white, dark, withegrid, darkgrid, stick]

sns.relplot(tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')

plt.show()
```

```python
sns.set_theme(context='poster', style='whitegrid', palette='colorblind') # palette gives [deep, muted, bright, pastel, dark, colorblind]

sns.relplot(tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')

plt.show()
```

```python
flights_long = sns.load_dataset('flights')
flights_long
```

```python
flights = flights_long.pivot(index='year', columns='month', values='passengers')
flights
```

```python
sns.heatmap(data=flights)
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot()

sns.heatmap(data=flights)

plt.show()
```

```python
flights = flights_long.pivot(index='month', columns='year', values='passengers')

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot()

sns.heatmap(data=flights)

plt.show()
```

```python
flights = flights_long.pivot(index='month', columns='year', values='passengers')

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot()

sns.heatmap(data=flights, annot=True, fmt='d')

plt.show()
```

```python
flights = flights_long.pivot(index='month', columns='year', values='passengers')

fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot()

sns.heatmap(data=flights, annot=True, fmt='d', linewidths=0.5)

plt.show()
```
